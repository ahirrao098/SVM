```python
data5=data4.drop(data4.columns[[6,64]],axis=1)
data5.shape
```

```python
data5.head()
```

```python
#balancing the data
```

```python
X = data5.drop('Salary',axis=1)
Y = data5['Salary']
X
```

```python
from sklearn.utils import resample
```

```python
train_data = pd.concat([X, Y], axis=1)

# separate minority and majority classes
negative = train_data[train_data.Salary==0]
positive = train_data[train_data.Salary==1]
```

```python
# upsample minority     # sample with replacement  # match number in majority class
pos_upsampled = resample(positive,replace=True,  n_samples=len(negative), random_state=27) # reproducible results

pos_upsampled;
```

```python
# combine majority and upsampled minority
upsampled = pd.concat([negative, pos_upsampled])
upsampled;
```

```python
# check new class counts
upsampled.Salary.value_counts()
```

```python
upsampled.head()
```

```python
data6=upsampled.copy()
data6.shape
```

```python
#Normalising
from sklearn import preprocessing
# Get column names first
names = data6.columns
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
scaled_data = scaler.fit_transform(data6)
scaled_data = pd.DataFrame(scaled_data, columns=names)
scaled_data
```
